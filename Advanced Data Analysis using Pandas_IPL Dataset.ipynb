#### 6.Consider the following dataframe. Reshape it such that it contains three columns: time, part (part_0_hits or part_1_hits) and hits (values corresponding to the part_0_hits and part_1_hits). Store it in part_df_melted

```python
d = {'time':[0,1,2], 'part_0_hits': [100,200,300], 'part_1_hits': [25,50,75]}
part_df = pd.DataFrame(d)
part_df
part_df_melted = part_df.set_index('time').stack().reset_index(name = 'hits').rename(columns = {'level_1': 'part'})
part_df_melted
```

#### 7. Use the part_df_melted generated in qn 6 and change the part column to show only the corresponding part numbers (0 or 1 inplace of part_0_hits or part_1_hits)

```python
part_df_melted['part'] = part_df_melted['part'].str.extract('(\d+)', expand = False)
part_df_melted
```

#### 8. Pivot the following dataframe to create 5 columns: labels, A,B,C,D.

#### 9. Pivot the following dataframe to create df_pivoted with four columns: category, A, B, C. Aggregate the values on sum.

```python
df = pd.DataFrame({'Name' : ['one', 'one', 'two', 'three'] * 3,
                   'Category' : ['A', 'B', 'C'] * 4,
               'Value' : range(12)})
df
```

```python
df_pivoted = pd.pivot_table(df, index='Name',columns='Category', values = 'Value', aggfunc='sum')
df_pivoted
```
